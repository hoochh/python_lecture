# 1. 네이버 API 뉴스 검색 및 저장

```python
import urllib.request
import datetime
import time
import json
import csv
import pandas as pd

client_id = "5jzCiWVGf6GeodQ9KwUV"
client_secret = "ySMUelE3jF"
```

```python
def getRequestUrl(url):
    req=urllib.request.Request(url)
    req.add_header('X-Naver-Client-Id',client_id)
    req.add_header('X-Naver-Client-Secret',client_secret)
    
    try:
        response=urllib.request.urlopen(req)
        if response.getcode()==200:
            print("[%s] Url Request Success"%datetime.datetime.now())
            return response.read().decode('utf-8')
    except Exception as e:
        print(e)
    print("[%s] Error for Url : %s"%(datetime.datetime.now(),url))
    return None # 1만 건 이상 요청할 시 error 발생
```

```python
def getNaverSearch(node, srcText, start, display):
    base="https://openapi.naver.com/v1/search"
    node="/%s.json"%node
    parameters="?query=%s&start=%s&display=%s"%(urllib.parse.quote(srcText),start,display)
    url=base+node+parameters
    responseDecode=getRequestUrl(url)
    
    if responseDecode==None:
        return None
    else:
        return json.loads(responseDecode)
```

```python
def getPostData(item, result):
    title=item['title'].strip().replace("\n","").replace("\r","").replace("&quot;","").replace("<b>","").replace("</b>","")
    
    #뉴스 검색 결과를 result에 저장
    result.append(title)
    return
```

```python
def main():
    node='news'
    srcText=input('검색어를 입력해주세요')
    cnt=0;
    result=[]
    
    jsonResponse=getNaverSearch(node, srcText, 1, 10)
    
    while((jsonResponse != None) and (jsonResponse['display'] != 0)):
        # 데이터 저장
        for item in jsonResponse['items']:
            cnt+=1
            getPostData(item, result)
            
            start=jsonResponse['start']+jsonResponse['display']
            jsonResponse=getNaverSearch(node, srcText, start, 10)
            
            # DataFrame 작성
            df=pd.DataFrame(result,columns=['title'])
            
        #print(result)
        
    df.to_csv('data/naver_news.csv', index=False)
    df
```

```python
main()
```

# 2. 형태소 분석

```python
from collections import Counter
from konlpy.tag import Okt
import pytagcloud
```

```python
news=pd.read_csv('data/naver_news.csv',encoding='utf-8')
news_title=news['title']
n_list=news_title.values
print(n_list)
```

```python
# 명사 추출
npl=Okt()
nouns=[]
for n in n_list:
    nouns=nouns+npl.nouns(n)
print(nouns)
```

```python
# 단어 집계 
count=Counter(nouns) 
print(count)
```

```python
tag=count.most_common(20)

taglist=pytagcloud.make_tags(tag,maxsize=80)

print(taglist)
```

# 3. WordCloud 생성

```python
pytagcloud.create_tag_image(taglist, 'news_title_wordcloud.jpg', size=(600, 400), fontname='Korean', rectangular=False)
```
