### <데이터 모델링>

```python
import pandas as pd
pd.set_option('mode.chained_assignment', None) #warning 방지용
import numpy as np

data=pd.read_csv('data/건강보험심사평가원_요양기관 개설 현황.csv',index_col=0,encoding='CP949',engine='python')

data.head()
```

```python
data['시도명'].unique()
```

```python
data['시군구명'].unique()
```

```python
addr=pd.DataFrame(data['시군구명'].apply(lambda v: v.split()[:1]).tolist(),columns=['군구명'])
addr.head()
```

```python
addr['시도명']=data['시도명'].tolist()
addr.head()
```

#### - 시도명, 시군구명 컬럼 결합

```python
addr['시도군구명']=addr.apply(lambda r:r['시도명']+' '+r['군구명'], axis=1)
# x축에 컬럼 추가
addr.head()
```

#### - 의료기관수 합계를 저장할 컬럼 생성

```python
addr['count']=0 #의료기관수 합계를 저장할 컬럼 만들기
addr.head()
```

#### - 시도군구명을 기준으로 그룹 생성 -> 그룹별 의료기관수 합계

```python
addr_group=pd.DataFrame(addr.groupby(['시도명','군구명','시도군구명'], as_index=False).count())
addr_group.head()
```

```python
addr_group=addr_group.set_index('시도군구명')
addr_group.head()
```

#### - 행정구역별 인수구 컬럼 추가

```python
population=pd.read_excel('data/행정구역_시군구_별__성별_인구수.xlsx')
population.head()
```

```python
population=population.rename(columns={'행정구역(시군구)별(1)':'시도명','행정구역(시군구)별(2)':'군구명'}) # 컬럼명 변경
population.head()
```

```python
#'군구' 컬럼에서 공백 제거하기
for element in range(0,len(population)):
    population['군구명'][element]=population['군구명'][element].strip()
    print(population['군구명'][element])
```

```python
#'시도'와 '군구'를 연결하여 '시도군구' 컬럼 추가
population['시도군구명']=population.apply(lambda r:r['시도명']+' '+r['군구명'],axis=1)
population.head()
```

```python
# 필요없는 소계 행 삭제
population=population[population.군구명 !='소계']
population.head()
```

```python
population=population.set_index("시도군구명")
population.head()
```

#### - 데이터프레임 병합

```python
addr_population_merge=pd.merge(addr_group, population, how='inner', left_index=True, right_index=True)
# left_index / right_index : 인덱스 표기 기능 없으면 인덱스 사라짐
addr_population_merge.head()
```

```python
# 필요한 컬럼만 추출
local_MC_Population=addr_population_merge[['시도명_x','군구명_x','count','총인구수 (명)']]
local_MC_Population.head()
```

```python
#컬럼명 변경
local_MC_Population=local_MC_Population.rename(columns={'시도명_x':'시도명','군구명_x':'군구명',
                                                       '총인구수 (명)':'인구수'})
local_MC_Population.head()
```

#### - 시도군구의 인구대비 의료기관수 비율

```python
MC_count=local_MC_Population['count']
local_MC_Population['MC_ratio']=MC_count.div(local_MC_Population['인구수'],axis=0)*100000
local_MC_Population.head()
```

#### - 블록맵 경계 좌표

```python
BORDER_LINES = [
    [(3, 2), (5, 2), (5, 3), (9, 3), (9, 1)], # 인천
    [(2, 5), (3, 5), (3, 4), (8, 4), (8, 7), (7, 7), (7, 9), (4, 9), (4, 7), (1, 7)], # 서울
    [(1, 6), (1, 9), (3, 9), (3, 10), (8, 10), (8, 9),
     (9, 9), (9, 8), (10, 8), (10, 5), (9, 5), (9, 3)], # 경기도
    [(9, 12), (9, 10), (8, 10)], # 강원도
    [(10, 5), (11, 5), (11, 4), (12, 4), (12, 5), (13, 5),
     (13, 4), (14, 4), (14, 2)], # 충청남도
    [(11, 5), (12, 5), (12, 6), (15, 6), (15, 7), (13, 7),
     (13, 8), (11, 8), (11, 9), (10, 9), (10, 8)], # 충청북도
    [(14, 4), (15, 4), (15, 6)], # 대전시
    [(14, 7), (14, 9), (13, 9), (13, 11), (13, 13)], # 경상북도
    [(14, 8), (16, 8), (16, 10), (15, 10),
     (15, 11), (14, 11), (14, 12), (13, 12)], # 대구시
    [(15, 11), (16, 11), (16, 13)], # 울산시
    [(17, 1), (17, 3), (18, 3), (18, 6), (15, 6)], # 전라북도
    [(19, 2), (19, 4), (21, 4), (21, 3), (22, 3), (22, 2), (19, 2)], # 광주시
    [(18, 5), (20, 5), (20, 6)], # 전라남도
    [(16, 9), (18, 9), (18, 8), (19, 8), (19, 9), (20, 9), (20, 10)], # 부산시
]
```

```python
def draw_blockMap(blockedMap, targetData, title, color):
    whitelabelmin=(max(blockedMap[targetData])-min(blockedMap[targetData]))+0.25+min(blockedMap[targetData]) # 값의 범위
    
    datalabel=targetData
    vmin=min(blockedMap[targetData]) # count 중에서 최대 최소값
    vmax=max(blockedMap[targetData])
    mapdata=blockedMap.pivot(index='y',columns='x',values=targetData) # pivot은 count 값으로 mapdata 생성 y=열, x=행
    masked_mapdata=np.ma.masked_where(np.isnan(mapdata),mapdata) 
    # np의 ma.masked_where(조건)은 조건(np가 Nan이면)을 만족하면 0, 만족하지 않으면 원래 값
    # nan 값을 제외하기 위해 마스크 기능 사용
    
    plt.figure(figsize=(8,13)) # 그림 사이즈
    plt.title(title) # 그림 타이틀
    plt.pcolor(masked_mapdata,vmin=vmin,vmax=vmax,cmap=color,edgecolor='#aaaaaa',linewidth=0.5)
    # pcolor만 적어도 기본적으로 칼라맵으로 그림을 그려줌
    # 설정을 
    
    #지역이름 표시
    for idx, row in blockedMap.iterrows():
        annocolor='white' if row[targetData] > whitelabelmin else 'black'
        
        #광역시는 구 이름이 겹치는 경우가 많아서 시단위 이름도 같이 표시한다.(중구, 서구)
        if row['광역시도'].endswith('시') and not row['광역시도'].startswith('세종'):
            dispname='{}\n{}'.format(row['광역시도'][:2], row['행정구역'][:-1])
            if len(row['행정구역']) <= 2:
                dispname += row['행정구역'][-1]
        else:
            dispname=row['행정구역'][:-1]
            
        #서대문구, 서귀포시 같이 이름이 3자 이상인 경우에 작은 글자로 표시
        if len(dispname.splitlines()[-1])>= 3:
            fontsize, linespacing=9.5,1.5
        else:
            fontsize, linespacing=11, 1.2
        
        plt.annotate(dispname, (row['x']+0.5, row['y']+0.5), weight='bold',
                     fontsize=fontsize, ha='center', va='center', color=annocolor,
                     linespacing=linespacing)
        
    # 시도 경계 그리기
    for path in BORDER_LINES:
        ys, xs = zip(*path)
        plt.plot(xs,ys,c='black', lw=4)
    
    plt.gca().invert_yaxis() # plt.gca(): 축 객체를 가져옴
    # 수학적으로 좌표는 좌측 아래에서 0으로 시작하기 때문에 축을 뒤집어줘야 함
    #plt.gca().set_aspect(1)
    plt.axis('off')

    cb = plt.colorbar(shrink = .1, aspect = 10)
    cb.set_label(datalabel)

    plt.tight_layout() # 만들어진 그래프의 여백을 없앰
    plt.savefig('data/'+'blockMap_'+targetData+'.png')
    plt.show()
```

```python
draw_blockMap(data_draw_korea_MC_Population_all, 'count', '행정구역별 요양기관 수', 'Blues')
```

```python
draw_blockMap(data_draw_korea_MC_Population_all, 'MC_ratio', '행정구역별 인구수 대비 요양기관 비율', 'Reds')
```
