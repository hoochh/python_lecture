```python
# 리스트 슬라이싱

list1=[[1,11],[2,22],[3,33]]
```

```python
print(list1[1][1])
print(list1[0])
print(list1[:][0]) # 리스트 전체 행의 1열 출력....안 됨! 열 단위 슬라이싱 불가능
#print(list1[1,1]) # 오류 발생
```

```python
# 넘파이 배열 슬라이싱 

list2=[1,10,100,1000,2,20,200,2000,3,30,300,3000]
arr1=np.array(list1)
arr2=np.array(list2).reshape(3,4)

print(arr1)
print(arr1[1][1])
print(arr1[1,1]) # 리스트와 달리 배열은 [행,열] 형태로 값을 불러올 수 있음
print(arr1[:,0]) # 리스트 전체 행의 1열 출력....열 단위 슬라이싱 가능
print(arr1[0,:]) # 리스트 전체 열의 1행 출력....행 단위 슬라이싱 가능
print(arr1[::1])

print(arr2)
print(arr2[1][1])
print(arr2[1,1]) # 리스트와 달리 배열은 [행,열] 형태로 값을 불러올 수 있음
print(arr2[:,0]) # 리스트 전체 행의 1열 출력....열 단위 슬라이싱 가능
print(arr2[0,:]) # 리스트 전체 열의 1행 출력....행 단위 슬라이싱 가능
print(arr2[::2,1::2])
```

```python
b=[5,3,4,2,5,1,7,6,9]
arr_b=np.array(b).reshape(3,3)
arr_b
```

```python
print(np.sort(arr_b)) # 각 행을 오름차순으로 정렬
print(np.sort(arr_b,axis=1)) # 행기준
print(np.sort(arr_b,axis=0)) # 열기준

print(np.sort(arr_b,axis=0)[::-1,::]) # 열기준 내림차순 정렬. 행이 바뀌어야 하므로 
                                        #[::-1,::] 에서 열부분이 생략된 것
print(np.sort(arr_b,axis=1)[::, ::-1]) # 행기준 내림차순 정렬 
```

```python
# 특정 간격으로 np 배열을 만들기

print(np.linspace(0,100,5)) # 0~100 사이에 같은 간격으로 5 개의 값을 출력
print(np.linspace(0,100,5, retstep=True)) # step이 얼마인지 보여줌
print(np.linspace(0,100,5, endpoint=True))
print(np.linspace(0,100,5, endpoint=False)) # 마지막 값을 포함하지 않고 5등분
print(np.logspace(1,2,4))
```
