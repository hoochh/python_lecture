```python
# yes24 도서 검색 결과 스크래핑 후 dataFrame으로 표현/csv 저장/mysql 저장
import urllib.request as req1
import requests
from bs4 import BeautifulSoup
import pandas as pd
import pymysql
```

```python
url="http://www.yes24.com/Product/Search?domain=ALL&query=python"
res=req1.urlopen(url)
bs=BeautifulSoup(res,'html.parser')

books=bs.select('#yesSchList > li')
```

```python
def image_save(url,i):
    save_name='data/book_images/image'+str(i)+'.jpg'
    req1.urlretrieve(url,save_name)
    print(save_name,'가 저장됨')
```

```python
book_list=[]
i=0
for book in books:
    title=book.select('div > div.item_info > div.info_row.info_name > a.gd_name')
    author=book.select('div > div.item_info > div.info_row.info_pubGrp > span.authPub.info_auth > a')
    publisher=book.select('div > div.item_info > div.info_row.info_pubGrp > span.authPub.info_pub > a')
    date=book.select('div > div.item_info > div.info_row.info_pubGrp > span.authPub.info_date')
    price=book.select('div > div.item_info > div.info_row.info_price > strong > em')
    
    print(title[0].text,author[0].text,publisher[0].text,date[0].text,price[0].text)
    
    b_list=(title[0].text,author[0].text,publisher[0].text,date[0].text,price[0].text)
    # tuple로 저장할 경우 db에 저장 수월함
    
    book_list.append(b_list)
    i+=1
    img=book.select('div > div.item_img > div.img_canvas > span > span > a > em > img')
    img_src=img[0]['data-original']
    # data-original에 실제 src 값이 들어 있음.
    # data-original이 있어야 data-original을 찾아 그 정보값을 쓸 수 있음
    
    # 아래 코드로 해당 class 이름을 가진 요소의 경로를 표시 -> data-original 확인 가능
    # book.find(class_=img_bdr).img
    # print(book)
    
    image_save(img_src,i)
    
print(book_list)
```

```python
# mysql에 데이터 저장
conn=pymysql.connect(host='localhost',user='lyk',password='1234',db='pydb',charset='utf8')
cursor=conn.cursor()
sql="insert into yes_book(title,author,publisher,date,price) values(%s,%s,%s,%s,%s)"
cursor.executemany(sql,book_list)
conn.commit()
conn.close()
```

```python
df=pd.DataFrame(book_list,columns=['title','author','publisher','date','price'])
df.to_csv('data/books.csv', index=False)
df
```
