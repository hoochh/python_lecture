```python
#매개변수를 추가하여 인터넷 리소스를 요청하는 방법 
import urllib.request
import urllib.parse

API="http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp" 
# 지역번호 : 전국 108, 서울/경기 109, 강원 105, 충북 131 충남 133, 
# 전북 146, 전남 156, 경북 143, 경남 159, 제주 184

values={'stnId':'105'} # stnId=105 와 같은 형식으로 출력됨
params=urllib.parse.urlencode(values) # 지역 번호 변환
print(params)

#요청 전용 url 생성
url=API+"?"+params 
print("url=",url)

#내용 가져와 출력
data=urllib.request.urlopen(url).read() 
text=data.decode('utf-8')
print(text)
```

```python
import requests

r = requests.request('get', 'http://unico2013.dothome.co.kr/crawling/exam.html')
r.encoding='utf-8'
print(type(r))
if r.text:
    print(r.text)
else:
    print('응답된 콘텐츠가 없어요.')
print('----------------------------')
r = requests.request('head', 'http://unico2013.dothome.co.kr/crawling/exam.html') 
r.encoding = 'utf-8'
print(type(r))
if r.text :
    print(r.text) 
else :
    print('응답된 콘텐츠가  없어요')
print('----------------------------')
r = requests.request('post', 'http://unico2013.dothome.co.kr/crawling/post.php', data= {'name':'백도', 'age' : 12})
r.encoding = 'utf-8'
print(type(r)) 
if r.text :
    print(r.text) 
else :
    print('응답된 콘텐츠가  없어요')
```
