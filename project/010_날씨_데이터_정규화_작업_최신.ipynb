# 1. 풍속 저장

```python
def wind_index(x):
    if x>11:
        return 5
    elif x>8:
        return 4
    elif x>5:
        return 3
    elif x>2:
        return 2
    return 1
```

```python
total_wind_df = pd.DataFrame()
day1_ago_wind_df = pd.DataFrame()

for idx, dong in tqdm(dong_df.iterrows()):
    dong_pd = pd.read_csv('data/weather_hour_0703/%s_시간별날씨_202101_202112.csv'%dong['dong'])
    
    wind_max = dong_pd.groupby(['date','region_id'])[['wind_speed']].max()
    wind_min = dong_pd.groupby(['date','region_id'])[['wind_speed']].min()
    wind_df = dong_pd.groupby(['date','region_id'])[['wind_speed']].mean()
    
    # 표준화 작업
    wind_std = StandardScaler().fit_transform(wind_df[['wind_speed']])
    wind_max_std =StandardScaler().fit_transform(wind_max[['wind_speed']])
    wind_min_std =StandardScaler().fit_transform(wind_min[['wind_speed']])
    
    
    # 하루 풍속의 평균, 최대, 최소 값 저장
    wind_df.columns=['wind_avg']
    wind_df['wind_max']=wind_max
    wind_df['wind_min']=wind_min
    
    # 하루 풍속의 평균, 최대, 최소의 표준화된 값 저장
    wind_df['wind_std']=wind_std
    wind_df['wind_max_std']=wind_max_std
    wind_df['wind_min_std']=wind_min_std
    
    # 하루 풍속의 평균, 최대 최소의 인덱스 지정
    wind_df['wind_avg_idx']=wind_df['wind_avg'].apply(lambda x:wind_index(x))
    wind_df['wind_max_idx']=wind_df['wind_max'].apply(lambda x:wind_index(x))
    wind_df['wind_min_idx']=wind_df['wind_min'].apply(lambda x:wind_index(x))

    ago_df = wind_df.reset_index()
    
    # 어제 풍속의 평균, 최대, 최소 데이터 프레임을 새로 작성 (수정될 가능성)
    ago_df['date'] = pd.to_datetime(ago_df['date'])
    ago_df['date'] = ago_df['date'].apply(lambda x : x - timedelta(days=1))
    
    total_wind_df = pd.concat([total_wind_df, wind_df],axis=0)
    day1_ago_wind_df = pd.concat([day1_ago_wind_df, ago_df],axis=0)
    
total_wind_df.to_csv('data/데이터원소화/전체날씨데이터/일일_풍속_평균_최대_최소_표준.csv')
day1_ago_wind_df.to_csv('data/데이터원소화/전체날씨데이터/어제_풍속_평균_최대_최소_표준.csv',index=False)
```

# 2. 습도 저장

```python
def hum_index(x):
    if x>90:
        return 5
    elif x>80:
        return 4
    elif x>70:
        return 3
    elif x>60:
        return 2
    return 1
```

```python
total_hum_df = pd.DataFrame()
day1_ago_hum_df = pd.DataFrame()

for idx, dong in tqdm(dong_df.iterrows()):
    dong_pd = pd.read_csv('data/weather_hour_0703/%s_시간별날씨_202101_202112.csv'%dong['dong'])
    
    hum_max = dong_pd.groupby(['date','region_id'])[['humidity']].max()
    hum_min = dong_pd.groupby(['date','region_id'])[['humidity']].min()
    
    hum_df = dong_pd.groupby(['date','region_id'])[['humidity']].mean()
    
    # 표준화 작업
    hum_std = StandardScaler().fit_transform(hum_df[['humidity']])
    hum_max_std =StandardScaler().fit_transform(hum_max[['humidity']])
    hum_min_std =StandardScaler().fit_transform(hum_min[['humidity']])
    
    # 습도의 평균, 최대, 최소 값 저장
    hum_df.columns=['hum_avg']
    hum_df['hum_max']=hum_max
    hum_df['hum_min']=hum_min
    
    # 습도의 평균, 최대, 최소 값 표준화된 값 저장
    hum_df['hum_std']=hum_std
    hum_df['hum_max_std']=hum_max_std
    hum_df['hum_min_std']=hum_min_std
    
    #습도의 평균, 최대, 최소 값 인덱스 지정
    # 지금 이건 2로 나눴을 때 몫을 저장한 것임(위에 적용함수 있음)
    hum_df['hum_avg_idx']=hum_df['hum_avg'].apply(lambda x:hum_index(x))
    hum_df['hum_max_idx']=hum_df['hum_max'].apply(lambda x:int(x)//2)
    hum_df['hum_min_idx']=hum_df['hum_min'].apply(lambda x:int(x)//2)

    ago_df = hum_df.reset_index()
    
    ago_df['date'] = pd.to_datetime(ago_df['date'])
    ago_df['date'] = ago_df['date'].apply(lambda x : x - timedelta(days=1))
    
    total_hum_df = pd.concat([total_hum_df, hum_df],axis=0)
    day1_ago_hum_df = pd.concat([day1_ago_hum_df, ago_df],axis=0)
    
total_hum_df.to_csv('data/데이터원소화/전체날씨데이터/일일_습도_평균_최대_최소_표준.csv')
day1_ago_hum_df.to_csv('data/데이터원소화/전체날씨데이터/어제_습도_평균_최대_최소_표준.csv',index=False)
```

# 3.기온 저장

```python
def temp_index(x):
    if x>23.5:
        return 5
    elif x>19.5:
        return 4
    elif x>15.5:
        return 3
    elif x>11.5:
        return 2
    return 1
```

```python
# 일교차를 위한 인덱스 함수
def chg_index(x):
    if x>12:
        return 5
    elif x>10:
        return 4
    elif x>8:
        return 3
    elif x>6:
        return 2
    return 1
```

```python
total_df = pd.DataFrame()
day1_ago_df = pd.DataFrame()

for idx, dong in tqdm(dong_df.iterrows()):
    dong_pd = pd.read_csv('data/weather_hour_0703/%s_시간별날씨_202101_202112.csv'%dong['dong'])
    
    max_df = dong_pd.groupby(['date','region_id'])[['temperature']].max()
    min_df = dong_pd.groupby(['date','region_id'])[['temperature']].min()
    
    df = dong_pd.groupby(['date','region_id'])[['temperature']].mean()
    
    #표준화 작업
    std_df = StandardScaler().fit_transform(df[['temperature']])
    max_std =StandardScaler().fit_transform(max_df[['temperature']])
    min_std =StandardScaler().fit_transform(min_df[['temperature']])
    
    # 하루 기온의 평균, 최대, 최소 값 저장
    df.columns=['temp_avg']
    df['temp_max']=max_df
    df['temp_min']=min_df
    
    #하루 기온의 평균, 최대,최소 표준화 된 값저장
    df['temp_std']=std_df
    df['temp_max_std']=max_std
    df['temp_min_std']=min_std
    
    # 인덱스 지정(마음대로 설정가능)
    df['temp_avg_idx']=df['temp_avg'].apply(lambda x:temp_index(x))
    df['temp_max_idx']=df['temp_max'].apply(lambda x:temp_index(x))
    df['temp_min_idx']=df['temp_min'].apply(lambda x:temp_index(x))

    ago_df = df.reset_index()
    
    ago_df['date'] = pd.to_datetime(ago_df['date'])
    ago_df['date'] = ago_df['date'].apply(lambda x : x - timedelta(days=1))
    
    total_df = pd.concat([total_df, df],axis=0)
    day1_ago_df = pd.concat([day1_ago_df, ago_df],axis=0)
    
total_df.to_csv('data/데이터원소화/전체날씨데이터/일일_기온_평균_최대_최소_표준.csv')
day1_ago_df.to_csv('data/데이터원소화/전체날씨데이터/어제_기온_평균_최대_최소_표준.csv',index=False)
```

```python
dong_pd = pd.read_csv('data/데이터원소화/전체날씨데이터/일일_기온_평균_최대_최소_표준.csv')

#기간 지정 : range(n,m) => n일 이상 m일 미만
#day_range = range(5,6)
day_range = [7,14,30]

for day in tqdm(day_range): # 기간 만큼 컬럼 추가
    
    max_avg_temp = [] #평균 기온 높았던 날
    min_avg_temp = [] #평균 기온 낮았던 날
    
    max_chg_temp = [] #일교차가 심했던 날
    
    max_temp = [] # 기온이 가장 높았던 날
    min_temp = [] # 기온이 가장 낮았던 날
    
    dong_pd['temp_chg']=dong_pd.apply(lambda x: x['temp_max']-x['temp_min'],axis=1)
    
    for _idx, _dong in dong_pd.iterrows():
        
        # 이전 날이 없는 경우 0 삽입되는 것
        if _idx-day<0:
            max_avg_temp.append(0.0)
            min_avg_temp.append(0.0)
            max_chg_temp.append(0.0)
            max_temp.append(0.0)
            min_temp.append(0.0)
            continue
            
        else:
            max_avg_temp.append(dong_pd.iloc[_idx-day:_idx+1,2].max()) #평균 기온 가장 높았던 날
            min_avg_temp.append(dong_pd.iloc[_idx-day:_idx+1,2].min()) #평균 기온 가장 낮았던 날
            
            max_chg_temp.append(dong_pd.iloc[_idx-day:_idx+1,-1].max()) # 일교차가 심했던 날
            
            max_temp.append(dong_pd.iloc[_idx-day:_idx+1,3].max()) # 기온이 가장 높았던 날의 최고 기온
            min_temp.append(dong_pd.iloc[_idx-day:_idx+1,4].min()) # 기온이 가장 낮았던 날의 최저 기온
            
    # 기온 값 컬럼의 이름을 지정
    temp_avg_max_col = 'temp_avg_max_%sd'%day # ex) tmep_avg_max_5d
    temp_avg_min_col = 'temp_avg_min_%sd'%day
    temp_chg_max_col = 'temp_chg_max_%sd'%day
    temp_max_col = 'temp_max_%sd'%day
    temp_min_col = 'temp_min_%sd'%day
    
    # 기온 값 인덱스 컬럼의 이름을 지정
    avg_max_idx_col='temp_avg_max_%sd_idx'%day
    avg_min_idx_col='temp_avg_min_%sd_idx'%day
    chg_max_idx_col='temp_chg_max_%sd_idx'%day
    max_idx_col='temp_max_%sd_idx'%day
    min_idx_col='temp_min_%sd_idx'%day
    
    dong_pd[temp_avg_max_col]=max_avg_temp
    dong_pd[temp_avg_min_col]=min_avg_temp
    dong_pd[temp_chg_max_col]=max_chg_temp
    dong_pd[temp_max_col]=max_temp
    dong_pd[temp_min_col]=min_temp

    #인덱스 지정
    # dong_pd[day_idx_col]=dong_pd[day_col].apply(lambda x:aqua_index_day(x))
    
    # dong_pd[avg_max_idx_col]=dong_pd[temp_avg_max_col].apply(lambda x:temp_index(x))
    # dong_pd[avg_min_idx_col]=dong_pd[temp_avg_min_col].apply(lambda x:temp_index(x))
    # dong_pd[chg_max_idx_col]=dong_pd[temp_chg_max_col].apply(lambda x:chg_index(x))
    # dong_pd[max_idx_col]=dong_pd[temp_max_col].apply(lambda x:temp_index(x))
    # dong_pd[min_idx_col]=dong_pd[temp_min_col].apply(lambda x:temp_index(x))
    
    dong_pd[avg_max_idx_col]=dong_pd[temp_avg_max_col].apply(lambda x:int(x)//2)
    dong_pd[avg_min_idx_col]=dong_pd[temp_avg_min_col].apply(lambda x:int(x)//2)
    dong_pd[chg_max_idx_col]=dong_pd[temp_chg_max_col].apply(lambda x:int(x)//2)
    dong_pd[max_idx_col]=dong_pd[temp_max_col].apply(lambda x:int(x)//2)
    dong_pd[min_idx_col]=dong_pd[temp_min_col].apply(lambda x:int(x)//2)
    
    print('%s일간의 기온 데이터 저장 완료'%day)
    
file_name='data/데이터원소화/전체날씨데이터/7_14_30일간_기온_평균최대_평균최소_최대일교차_최대기온_최소기온_index.csv'
dong_pd.to_csv(file_name,index=False,float_format='%.2f')

dong_pd
```

# 4. 강수량 저장

```python
def aqua_index(x):
    # if x>3:
    #     return 3
    if x>0 and x<=10:
        return x//1
    elif x>10:
        return x//10
    return 0
```

```python
def aqua_index_day(x):
    if x>50:
        return 10
    elif x>45:
        return 9
    elif x>40:
        return 8
    elif x>35:
        return 7
    elif x>30:
        return 6
    elif x>25:
        return 5
    elif x>20:
        return 4
    elif x>10:
        return 3
    elif x>5:
        return 2
    elif x>0:
        return 1
    return 0
```

```python
def aqua_index_std(x,mean,std):
    if mean+(2*std)<x:
        return 5
    elif mean+std<x:
        return 4
    elif mean-(2*std)>x:
        return 1
    elif mean-std>x:
        return 2
    else:
        return 3
```

```python
total_df = pd.DataFrame()
day1_ago_df = pd.DataFrame()

for idx, dong in tqdm(dong_df.iterrows()):
    dong_pd = pd.read_csv('data/weather_hour_0703/%s_시간별날씨_202101_202112.csv'%dong['dong'])
    
    dong_pd['precipitation'] = dong_pd['precipitation'].apply(lambda x : x if x>-1 else 0)
    
    max_df = dong_pd.groupby(['date','region_id'])[['precipitation']].max()
    min_df = dong_pd.groupby(['date','region_id'])[['precipitation']].min()
    
    df = dong_pd.groupby(['date','region_id'])[['precipitation']].sum()
    
    std_df = StandardScaler().fit_transform(df[['precipitation']])
    max_std =StandardScaler().fit_transform(max_df[['precipitation']])
    min_std =StandardScaler().fit_transform(min_df[['precipitation']])
    
    df.columns=['aqua_sum']
    df['aqua_max']=max_df
    df['aqua_min']=min_df
    
    df['aqua_std']=std_df
    df['aqua_max_std']=max_std
    df['aqua_min_std']=min_std
    
    
    # 이거는 정규 분포를 적용하였음
    df['aqua_sum_idx']=df['aqua_sum'].apply(lambda x:aqua_index_std(x, std_df.mean(), std_df.std())) #<=이거 aqua_std(강수량값, 평균, 표준편차)
    df['aqua_max_idx']=df['aqua_max'].apply(lambda x:aqua_index_std(x, max_std.mean(), max_std.std()))
    df['aqua_min_idx']=df['aqua_min'].apply(lambda x:aqua_index_std(x, max_std.mean(), max_std.std()))
    # df['aqua_sum_idx']=df['aqua_sum'].apply(lambda x:aqua_index(x))
    # df['aqua_max_idx']=df['aqua_max'].apply(lambda x:aqua_index(x))
    # df['aqua_min_idx']=df['aqua_min'].apply(lambda x:aqua_index(x))

    ago_df = df.reset_index()
    
    ago_df['date'] = pd.to_datetime(ago_df['date'])
    ago_df['date'] = ago_df['date'].apply(lambda x : x - timedelta(days=1))
    
    total_df = pd.concat([total_df, df],axis=0)
    day1_ago_df = pd.concat([day1_ago_df, ago_df],axis=0)
    
total_df.to_csv('data/데이터원소화/전체날씨데이터/일일_강수량_평균_최대_최소_표준.csv')
day1_ago_df.to_csv('data/데이터원소화/전체날씨데이터/어제_강수량_평균_최대_최소_표준.csv',index=False)
```
