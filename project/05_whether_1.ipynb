```python
import pandas as pd

for idx, dong in enumerate(dong_list):
    dong_pd = pd.read_csv(r'data/날씨데이터 1_20/%s_강수_202101_202112.csv'%(dong))
    humidity= pd.read_csv(r'data/날씨데이터 1_20/%s_습도_202101_202112.csv'%(dong)).iloc[:,-1]
    temperature= pd.read_csv(r'data/날씨데이터 1_20/%s_기온_202101_202112.csv'%(dong)).iloc[:,-1]
    wind_speed = pd.read_csv(r'data/날씨데이터 1_20/%s_풍속_202101_202112.csv'%(dong)).iloc[:,-1]

    dong_pd.columns=['day','hour','precipitation']

    # dong_pd['city_id']=city_code[idx]
    dong_pd['humidity']=humidity
    dong_pd['wind_speed']=wind_speed
    dong_pd['temperature']=temperature
    dong_pd['dong']=dong
    dong_pd['day'] = dong_pd['day'].apply(lambda x : x.strip())
    dong_pd =pd.DataFrame([ x for i,x in dong_pd.iterrows() if len(x['day'])<3 ])
    
    month=0
    date_list=[]
    for i, val in dong_pd.iterrows():
        if val['day']=='1' and val['hour']==0.0:
            month+=1
        date='2021.%s.%s'%(month,val['day'])
        date_list.append(date)

    dong_pd['date']=date_list
    del dong_pd['day']

    dong_pd.to_csv('data/wheather_hour/%s_시간별날씨_202101_202112.csv'%dong, index=False)
    print('%s ::: save %s wheather data'%(idx, dong))
```

```python
borders = pd.read_csv('data/update_city_map.csv')

for idx, dong in enumerate(dong_list):
    dong_pd = pd.read_csv('data/wheather_hour/%s_시간별날씨_202101_202112.csv'%dong)
    print(dong_pd[dong_pd['wind_speed']==-1],'\n===============================================================================')
    
    idxs = dong_pd[dong_pd['wind_speed']==-1].index
    
    for i in idxs:
        print(dong_pd.iloc[[i-1,i+1]])
        d1 = dong_pd.iloc[[i-1],3]
        d2 = dong_pd.iloc[[i+1],3]
        
        if dong_pd.iloc[[i-1],3].any()<0:
            d1 = dong_pd.iloc[[i-2],3]
        if dong_pd.iloc[[i+1],3].any()<0:
            d2 = dong_pd.iloc[[i+2],3]
            
        avg = (list(d1)[0]+list(d2)[0])/2
        dong_pd.iloc[i,3]=avg

        print('mean 1ago 1after ::',dong_pd.iloc[i],'\n')
    
    dong_pd = dong_pd.drop(dong_pd[dong_pd['humidity']==-1].index)
    
    
    dong_pd['date'] = pd.to_datetime(dong_pd['date'])
    avg_data = dong_pd.groupby(['date'],as_index=False)[['humidity','wind_speed','temperature']].mean()
    sum_data = dong_pd.groupby(['date'],as_index=False)['precipitation'].sum()
    dong_pd = pd.merge(avg_data,sum_data,  how='inner')
    
    dong_pd['region_id'] = list(borders[borders['ADM_DR_NM']==dong]['ADM_DR_CD'])[0]
    # dong_pd['region']=dong
    
    dong_pd.to_csv('data/weather_day/%s_일일_날씨_202101_202112.csv'%dong,index=False,float_format='%.2f')
    print('%s_일일_날씨_202101_202112.csv 저장 완료!'%dong)
```
