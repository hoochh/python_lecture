```python
city = pd.read_csv('data/incheon_city.csv')

# print (city[city['dong']=='연수2동'])
dong_list = list(city['dong'])
len(dong_list)
# dong_list
# city[city['dong']=='일신동']
```

```python
import pandas as pd

for idx, dong in enumerate(dong_list):
    dong_pd = pd.read_csv('data/weather/%s_강수_202101_202112.csv'%(dong))
    humidity= pd.read_csv('data/weather/%s_습도_202101_202112.csv'%(dong)).iloc[:,-1]
    temperature= pd.read_csv('data/weather/%s_기온_202101_202112.csv'%(dong)).iloc[:,-1]
    wind_speed = pd.read_csv('data/weather/%s_풍속_202101_202112.csv'%(dong)).iloc[:,-1]

    dong_pd.columns=['day','hour','precipitation']

    # dong_pd['city_id']=city_code[idx]
    dong_pd['humidity']=humidity
    dong_pd['wind_speed']=wind_speed
    dong_pd['temperature']=temperature
    dong_pd['dong']=dong
    # day 공백 제거
    dong_pd['day'] = dong_pd['day'].apply(lambda x : x.strip())
    # 필요없는 문자열 제거
    dong_pd =pd.DataFrame([ x for i,x in dong_pd.iterrows() if len(x['day'])<3 ])
    
    month=0
    date_list=[]
    for i, val in dong_pd.iterrows():
        if val['day']=='1' and val['hour']==0.0:
            month+=1
        date='2021.%s.%s'%(month,val['day'])
        date_list.append(date)

    dong_pd['date']=date_list
    del dong_pd['day']

    dong_pd.to_csv('data/weather_hour/%s_시간별날씨_202101_202112.csv'%dong, index=False)
    print('%s ::: save %s wheather data'%(idx, dong))
```
