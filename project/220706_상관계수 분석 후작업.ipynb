```python
temp_df=pd.read_csv('data/데이터원소화/전체날씨데이터/7_14_30일간_기온_평균최대_평균최소_최대일교차_최대기온_최소기온.csv')
temp_df
```

```python
temp_df = temp_df.query('"2021-08-11"<= date <= "2021-10-26"')
temp_df
```

### 7일간 최대 일교차 10이상 12미만

```python
temp_ch_7d=[]
temp_ch_region=[]
temp_ch_date=[]
for idx, temp in temp_df.iterrows():
    if (temp['temp_chg_max_7d']>=10 and temp['temp_chg_max_7d']<12):
        temp_ch_7d.append(temp['temp_chg_max_7d'])
        temp_ch_region.append(temp['region_id'])
        temp_ch_date.append(temp['date'])

data = dict()
data['date']=temp_ch_date
data['region_id']=temp_ch_region
data['temp_chg_max_7d']=temp_ch_7d

temp_7d=pd.DataFrame(data)
temp_7d = temp_7d.set_index(['date','region_id'])
temp_7d
```

### 14일간 최대 일교차 10이상 12미만

```python
temp_ch_14d=[]
temp_ch_region_14d=[]
temp_ch_date_14d=[]
for idx, temp in temp_df.iterrows():
    if (temp['temp_chg_max_14d']>=10 and temp['temp_chg_max_14d']<12):
        temp_ch_14d.append(temp['temp_chg_max_14d'])
        temp_ch_region_14d.append(temp['region_id'])
        temp_ch_date_14d.append(temp['date'])

data = dict()
data['date']=temp_ch_date_14d
data['region_id']=temp_ch_region_14d
data['temp_chg_max_14d']=temp_ch_14d

temp_14d=pd.DataFrame(data)
temp_14d = temp_14d.set_index(['date','region_id'])
temp_14d
```

### 30일간 최대 일교차 10이상 12미만

```python
temp_ch_30d=[]
temp_ch_region_30d=[]
temp_ch_date_30d=[]
for idx, temp in temp_df.iterrows():
    if (temp['temp_chg_max_30d']>=10 and temp['temp_chg_max_30d']<12):
        temp_ch_30d.append(temp['temp_chg_max_30d'])
        temp_ch_region_30d.append(temp['region_id'])
        temp_ch_date_30d.append(temp['date'])

temp_30d=pd.DataFrame({'date':temp_ch_date_30d, 'region_id':temp_ch_region_30d, 'temp_chg_max_30d':temp_ch_30d})
temp_30d = temp_30d.set_index(['date','region_id'])
temp_30d
```

### 데이터 merge, 2021-08-11 ~ 2021-10-26 기간 설정

```python
temp_ch = pd.merge(temp_7d, temp_14d, how='outer', left_index=True, right_index=True)
temp_ch = pd.merge(temp_ch, temp_30d, how='outer', left_index=True, right_index=True)

temp_ch = temp_ch.reset_index()

temp_ch
```

```python
ch_index=pd.DataFrame()
ch_index['date']=temp_ch['date']
ch_index['region_id']=temp_ch['region_id']

ch_index
```
