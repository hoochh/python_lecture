```python
import pandas as pd
```

```python
city = pd.read_csv('data/incheon_city.csv')

# print (city[city['dong']=='연수2동'])
dong_list = list(city.iloc[47:47+50,:]['dong'])

dong_list
```

```python
city_code = list(city.iloc[47:47+50,:]['id'])
city_code
```

```python
import pandas as pd

for idx, dong in enumerate(dong_list):
    dong_pd = pd.read_csv(r'C:\kej\날씨데이터 21_40\%s\%s_강수_202101_202112.csv'%(dong,dong))
    humidity= pd.read_csv(r'C:\kej\날씨데이터 21_40\%s\%s_습도_202101_202112.csv'%(dong,dong)).iloc[:,-1]
    temperature= pd.read_csv(r'C:\kej\날씨데이터 21_40\%s\%s_기온_202101_202112.csv'%(dong,dong)).iloc[:,-1]
    wind_speed = pd.read_csv(r'C:\kej\날씨데이터 21_40\%s\%s_풍속_202101_202112.csv'%(dong,dong)).iloc[:,-1]

    dong_pd.columns=['day','hour','precipitation']

    # dong_pd['city_id']=city_code[idx]
    dong_pd['humidity']=humidity
    dong_pd['wind_speed']=wind_speed
    dong_pd['temperature']=temperature
    dong_pd['dong']=dong
    dong_pd['day'] = dong_pd['day'].apply(lambda x : x.strip())
    dong_pd =pd.DataFrame([ x for i,x in dong_pd.iterrows() if len(x['day'])<3 ])
    
    month=0
    date_list=[]
    for i, val in dong_pd.iterrows():
        if val['day']=='1' and val['hour']==0.0:
            month+=1
        date='2021.%s.%s'%(month,val['day'])
        date_list.append(date)

    dong_pd['date']=date_list
    del dong_pd['day']

    dong_pd.to_csv('data/wheather_hour/%s_시간별날씨_202101_202112.csv'%dong, index=False)
    print('%s ::: save %s wheather data'%(idx, dong))
```
