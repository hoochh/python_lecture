```python
border_pd=pd.read_csv('data/update_city_map.csv',encoding='utf-8')
border_pd
```

```python
flood_pd['dong']=flood_pd.iloc[:,1].apply(lambda v: v.split()[:1][0])
flood_pd
```

```python
flood_pd[flood_pd['dong']=='용현동']
```
