```python
import pandas as pd
```

```python
city = pd.read_csv('data/incheon_city.csv')
dong_list = list(city['dong'])

len(dong_list)
```

```python
borders = pd.read_csv('data/update_city_map.csv')
```

```python
for idx, dong in enumerate(dong_list):
    hum_max=[]
    dong_pd = pd.read_csv('data/weather_hour/%s_시간별날씨_202101_202112.csv'%dong)
    hum_max=dong_pd.groupby(['date'])[['humidity']].max()
    
    hum_max.to_csv('data/weather_hum/%s_일일_최고습도_202101_202112.csv'%dong,index=False,float_format='%.2f')
    print('%s_일일_최고습도_202101_202112.csv 저장 완료!'%dong)
```

```python
for idx, dong in enumerate(dong_list):
    dong_pd = pd.read_csv('data/weather_hour/%s_시간별날씨_202101_202112.csv'%dong)
    # print(dong_pd[dong_pd['wind_speed']==-1],'\n===============================================================================')

    # dong_pd['precipitation'].value_counts()
    # dong_pd['humidity'].unique()
    # dong_pd['wind_speed'].unique()
    # dong_pd['temperature'].unique()
    
    # dong_pd[dong_pd['humidity']==-1]
    
    idxs = dong_pd[dong_pd['wind_speed']==-1].index
    
    for i in idxs:
        # print(dong_pd.iloc[[i-1,i+1]])
        d1 = dong_pd.iloc[[i-1],3]
        d2 = dong_pd.iloc[[i+1],3]
        
        if dong_pd.iloc[[i-1],3].any()<0:
            d1 = dong_pd.iloc[[i-2],3]
        if dong_pd.iloc[[i+1],3].any()<0:
            d2 = dong_pd.iloc[[i+2],3]
            
        avg = (list(d1)[0]+list(d2)[0])/2
        dong_pd.iloc[i,3]=avg

        # print('mean 1ago 1after ::',dong_pd.iloc[i],'\n')
    
    dong_pd = dong_pd.drop(dong_pd[dong_pd['humidity']==-1].index)
    
    
    dong_pd['date'] = pd.to_datetime(dong_pd['date'])
    
    temp_max=dong_pd.groupby(['date'])[['temperature']].max()
    temp_min=dong_pd.groupby(['date'])[['temperature']].min()
    temp_avg=dong_pd.groupby(['date'])[['temperature']].mean()
    
    hum_avg=dong_pd.groupby(['date'])[['humidity']].mean()
    hum_min=dong_pd.groupby(['date'])[['humidity']].min()
    
    aqua_sum=dong_pd.groupby(['date'])['precipitation'].sum()
    
    dong_pd = pd.merge(temp_max,temp_min, left_index=True, right_index=True, how='inner')
    dong_pd = pd.merge(dong_pd,temp_avg, left_index=True, right_index=True, how='inner')
    dong_pd = pd.merge(dong_pd,hum_avg, left_index=True, right_index=True, how='inner')
    dong_pd = pd.merge(dong_pd,hum_min, left_index=True, right_index=True, how='inner')
    dong_pd = pd.merge(dong_pd,aqua_sum, left_index=True, right_index=True, how='inner')
    
    dong_pd=dong_pd.reset_index()
    dong_pd.columns=['date','temp_max','temp_min','temp_avg','hum_avg','hum_min','aqua_sum']
    
    dong_pd['region_id'] = list(borders[borders['ADM_DR_NM']==dong]['ADM_DR_CD'])[0]
    # dong_pd['region']=dong
    print(dong_pd)
    dong_pd.to_csv('data/weather_day/%s_일일_날씨_202101_202112.csv'%dong,index=False,float_format='%.2f')
    print('%s_3일_날씨_202101_202112.csv 저장 완료!'%dong)
```

```python
for idx, dong in enumerate(dong_list):
    dong_pd = pd.read_csv('data/weather_day/%s_일일_날씨_202101_202112.csv'%dong)
    aqua_3d_list=[]
    for _idx, _dong in dong_pd.iterrows():
        if _idx-3<0:
            aqua_3d_list.append(0.0)
            continue
        else:
            aqua_3d_list.append(dong_pd.iloc[_idx-3:_idx,6].sum())

    dong_pd['aqua_3d']=aqua_3d_list
    
    
    file_name='data/weather_3day/%s_3일간_날씨_202101_202112.csv'%str(dong)
    dong_pd.to_csv(file_name,index=False,float_format='%.2f')
    print('%s_3일간_날씨_202101_202112.csv 저장 완료!'%dong)
```
