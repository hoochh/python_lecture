# pairplot, lmplot, jointplot 활용

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style='darkgrid')
            
%matplotlib inline
```

```python
# titanic dataset load
titanic=sns.load_dataset('titanic')
print(titanic.shape)
titanic.head()
```

```python
# tips dataset load
tips=sns.load_dataset('tips')
print(tips.shape)
tips.head()
```

# pairplot
- 그리드(grid) 형태로 각 집합의 조합에 대해 히스토그램과 분포도를 표시
- 숫자형 column에 대해서만 그려준다
- 자기 자신과의 관계는 histogram으로 나타나며 다른 요소와의 관계가 분포도로 표시된다

```python
sns.pairplot(tips)
plt.show()
```

```python
# hue 옵션으로 구분해서 볼 수도 있다.
sns.pairplot(tips, hue='size')
plt.show()
```

# lmplot
- Implot은 column간의 선형관계를 확인하기에 용이한 차트
- 또한 outlier도 같이 짐작해 볼 수 있다.

```python
sns.lmplot(x='total_bill',y='tip',height=8,data=tips)
plt.show()
```

```python
sns.lmplot(x='total_bill',y='tip', hue='smoker',height=8,data=tips)
plt.show()
```

```python
# col_wrap으로 한 줄에 표기할 column의 갯수를 명시할 수 있다.
sns.lmplot(x='total_bill',y='tip', hue='smoker',col='day',col_wrap=2,height=6,data=tips)
plt.show()
```

# jointplot
- scatter(산점도)와 histogram(분포도)을 동시에 그려주며 숫자형 데이터만 표현 가능

```python
sns.jointplot(x='total_bill',y='tip',height=6,data=tips)
plt.show()
```

```python
# 선형관계를 표현하는 regression 라인 그리기도 가능하다
# 옵션에 kind='reg' 추가
sns.jointplot(x='total_bill',y='tip',height=6,data=tips,kind='reg')
plt.show()
```
