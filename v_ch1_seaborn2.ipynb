# Cat plot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style='whitegrid')
            
%matplotlib inline
```

```python
from matplotlib import rc, font_manager
# 한글 처리를 위해 폰트 설정
fontname="C:/Windows/Fonts/gulim.ttc"
font_name=font_manager.FontProperties(fname=fontname).get_name()

rc('font',family=font_name)

# 그림 크기
plt.rcParams['figure.figsize']=(12,9)
# 마이너스 표기
plt.rcParams['axes.unicode_minus']=False
```

```python
# 데이터 셋
tips=sns.load_dataset('tips')
tips.head()
```

```python
# 요일별 요금 정산 금액
sns.catplot(x='day',y='total_bill',data=tips)
```

```python
# 펼쳐서 겹침을 없애 줌, 정확한 개수를 알 수 있음
sns.catplot(x='day',y='total_bill',data=tips, kind='swarm')
```

```python
# 색상 설정
sns.catplot(x='day',y='total_bill',data=tips, kind='swarm', hue='sex')
```

```python
# 가운데 세로 줄로 균형
sns.catplot(x='day',y='total_bill',data=tips,kind='violin')
```

# Violin plot
- 바이올린처럼 생긴 차트
- column에 대한 데이터의 비교 분포도를 확인할 수 있다.
- 곡선진 부분은 데이터의 분포를 나타낸다.
- 양쪽 끝 뾰족한 부분은 데이터의 최소값과 최대값을 나타낸다.

```python
# 기본 violinplot 그리기
sns.violinplot(x=tips["total_bill"])
plt.show()
```

```python
# x, y축을 지정해줌으로써 바이올린을 분할하여 비교 분포를 볼 수 있다.
sns.violinplot(x='day',y='total_bill',data=tips)
plt.show()
```

```python
# violinplot의 하이라이트로 split 옵션으로 바이올린을 합쳐서 볼 수 있다.
# catplot의 kind='violin'과 유사한 결과를 얻을 수 있음
sns.violinplot(x='day',y='total_bill',hue='sex',data=tips,palette='muted',split=True)
plt.show()
```

```python
#sns.violinplot(x='day',y='total_bill',hue='smoker',data=tips,inner='stick',split=True)
sns.catplot(x='day',y='total_bill',hue='smoker',kind='violin',data=tips,inner='stick',split=True)
plt.show()
# split: 중앙선으로 카테고리 성별로 나누어짐
# inner='stick': 중앙에 선 그림(실제로 값이 있는 부분)
```

# Box plot
- 실제 값이 어디에 많이 몰려있는지를 보여줌

```python
# 박스 중앙이 50%
# 박스를 벗어난 값은 매우 빈도가 드문 값임
sns.catplot(x='day', y='total_bill', data=tips, kind='box')
```

```python
# 좀 더 상세한 값의 분포 구간을 보여줌
sns.catplot(x='day', y='total_bill', data=tips, kind='boxen')
```

```python
# 값의 분포가 어디에 있는지를 보여줌
sns.catplot(x='day', y='total_bill', data=tips, kind='point')
```

```python
# 세로선은 오차범위를 의미
sns.catplot(x='day', y='total_bill', data=tips, kind='bar')
```

```python
sns.catplot(x='day', data=tips, kind='count')
```
